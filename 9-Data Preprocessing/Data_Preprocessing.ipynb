```python
import numpy as np
import pandas as pd
```

Handaling Null Values

```python
data = [['Steve',25,55],['Linus',28,60],['Elon',26]]
df = pd.DataFrame(data=data,columns=['Name','Age','Weight'])
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna()
# The changes won't be reflected in the original dataframe until we set inplace=True
```

```python
df
```

Imputation

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(df[['Weight']])
df['Weight'] = imputer.transform(df[['Weight']])
df
```

```python
df.fillna(df.mean())
```

Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
X = std.fit_transform(df[['Age','Weight']])
X
```

**Handling Categorical Variables**

Ordinal Categorical Variables

```python
df_cat = pd.DataFrame(data = 
                     [['green','M',10.1,'class1'],
                      ['blue','L',20.1,'class2'],
                      ['white','M',30.1,'class1']])
df_cat.columns = ['color','size','price','classlabel']
df_cat
```

```python
size_mapping = {'M':1,'L':2}
df_cat['size'] = df_cat['size'].map(size_mapping)
```

```python
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
df_cat['classlabel'] = class_le.fit_transform(df_cat['classlabel'].values)
```

Nominal Categorical Variables

```python
# Using one-hot encoding
df_cat = pd.get_dummies(df_cat[['color','size','price']])
df_cat
```
